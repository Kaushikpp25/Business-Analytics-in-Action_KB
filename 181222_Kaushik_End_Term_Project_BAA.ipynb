```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
tel2=tel1.copy()
tel2.head(40)
```

```python
for i in tel2.columns[:-1]:
    tel2[i]=le.fit_transform(tel2[i])

tel2.head()
```

```python
tel2.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = OneHotEncoder(categorical_features = 'all')
```

```python
tel3 = onehotencoder.fit_transform(tel2).toarray()
```

```python
tel3=pd.DataFrame(tel3)
tel3.head(5)
```

```python
tel2_encoded=pd.DataFrame(tel3)
tel2_encoded.info()
```

```python
tel2_encoded['class']=tel1['Churn']
```

```python
tel2_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x=tel2_encoded.columns[:-1]
x_train,x_test,y_train,y_test=train_test_split(tel2_encoded[x],tel2_encoded['class'])
```

```python
print(x_train.shape,y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x= StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
print(y_train.shape,y_train.shape)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predicted_class=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,predicted_class)
accuracy
```

```python
accuracy1 = accuracy*100
print(accuracy1)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred=model.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm                    
```
